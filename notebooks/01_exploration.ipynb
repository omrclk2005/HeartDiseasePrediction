```python
import pandas as pd

df = pd.read_csv('../data/raw/heart.csv')

print("Original Dataset Size:", df.shape)

# Find and remove duplicate rows in the dataset.
df_clean = df.drop_duplicates()

print(f"Number of duplicate rows removed: {df.shape[0] - df_clean.shape[0]}")
print("Cleaned Dataset Size:", df_clean.shape)

missing_data = df.isnull().sum().sum()
print(f"Number of missing values: {missing_data}")

# One-Hot Encoding
columns_to_encode = ['cp', 'restecg', 'slope', 'thal']
df_processed = pd.get_dummies(df_clean, columns=columns_to_encode, prefix=columns_to_encode, dtype=int)

# Result
print("\nFirst 5 Rows of the Processed Dataset:")
print(df_processed.head())

# Saving the processed file
df_processed.to_csv('../data/processed/heart_processed.csv', index=False)
```
